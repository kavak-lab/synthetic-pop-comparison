```r
# This PUMF file (AAU9UOzK.csv) may not be available on github since it is quite large
raw_inds <- read.csv(paste0("AAU9UOzK.csv"),stringsAsFactors = FALSE)
```

```r
age_group = raw_inds['agegrp']
sex = raw_inds['sex']
ethnicity = raw_inds['ethder']
cma = raw_inds['cma']
id = raw_inds['ppsort']
```

```r
inds_all = cbind(id, cma, age_group, sex, ethnicity)
```

```r
head(inds_all)
```

```r
vancouver_cma = 933 #Filtering out vancouver only.
inds = data.frame(matrix(0, ncol = 5, nrow = 0))
names(inds) = names(inds_all)
v = 0
counter = 1
while(counter<=length(inds_all[[1]])){ #length(inds_all[[1]])
    if(inds_all[counter, 'cma']==vancouver_cma){
        inds = rbind(inds, inds_all[counter,])
        v = v+1
    }
    counter = counter+1
    #print(counter)
}
```
