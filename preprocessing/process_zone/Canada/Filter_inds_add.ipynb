```python
import pandas as pd
import numpy as np
```

```python
raw_inds = pd.read_csv("AAU9UOzK.csv")
cur_inds = pd.read_csv("../../../synthetic_data_DA_add/inds.csv")
cur_inds.head()
```

```python
raw_inds['mob1'].head()
```

```python
mar_status = []
dwelling = []
mobility = []
for index, row in raw_inds.iterrows():
    if row['ppsort'] in cur_inds['PPSORT'].to_list():
        if row['marsth']==6:
            mar_status.append('widowed')
        elif row['marsth']==2:
            mar_status.append('married')
        elif row['marsth']==3:
            mar_status.append('living.common.law')
        elif row['marsth']==4:
            mar_status.append('separated')
        elif row['marsth']==5:
            mar_status.append('divorced')
        else:
            mar_status.append('no.married')
        if row['dtype'] == 1:
            dwelling.append('single.detached.house')
        elif row['dtype'] == 2:
            dwelling.append('apartment')
        else:
            dwelling.append('other.attached')
        if row['mob1'] == 1:
            mobility.append('non.movers')
        elif row['mob1'] == 2:
            mobility.append('nonmigrants.movers')
        elif row['mob1'] == 5:
            mobility.append('interprovincial.movers')
        elif row['mob1'] == 6:
            mobility.append('external.movers')
        else:
            mobility.append('other.movers')
mar_status = pd.Series(mar_status)
dwelling = pd.Series(dwelling)
mobility = pd.Series(mobility)
cur_inds['MARST'] = mar_status
cur_inds['DWEL'] = dwelling
cur_inds['MOBL'] = mobility
```
