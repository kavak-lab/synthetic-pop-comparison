```python
import pandas as pd
import numpy as np
```

```python
cons_fairfax = pd.read_csv("../../../census_data/Processed/US/CT/cons_us_ct.csv")
synth_pop_fairfax = pd.read_csv("../../../synthetic_data/US/CT/synthetic_group_data_RPWR.csv")
synth_pop_fairfax.columns
```
